```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
dtypes={
    'elapsed_time':np.int32,
    'event_name':'category',
    'name':'category',
    'level':np.uint8,
    'room_coor_x':np.float32,
    'room_coor_y':np.float32,
    'screen_coor_x':np.float32,
    'screen_coor_y':np.float32,
    'hover_duration':np.float32,
    'text':'category',
    'fqid':'category',
    'room_fqid':'category',
    'text_fqid':'category',
    'fullscreen':'category',
    'hq':'category',
    'music':'category',
    'level_group':'category'}
# get 10000 rows from the training dataset
dataset_df = pd.read_csv('data/train.csv', dtype=dtypes, nrows=10000)
print("Full train dataset shape is {}".format(dataset_df.shape))
```

```python
#  data explore  
# 1. distribution of event_name, name, level
fig, ax = plt.subplots(1, 3, figsize=(16, 4))
dataset_df['event_name'].value_counts().plot(kind='bar', ax=ax[0], title='event_name')
dataset_df['name'].value_counts().plot(kind='bar', ax=ax[1], title='name')
dataset_df['level'].value_counts().plot(kind='bar', ax=ax[2], title='level')
plt.show()
```
